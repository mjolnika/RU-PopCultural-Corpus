### Собственно код

Подфункции:

0.1. Преобразуем строку с исполнителями в список

```python
def artistlist(artists_as_string):
    lines = artists_as_string.splitlines()
    artist_list = []
    for line in lines:
        if len(line)>1:
            artist_list.append(line)
    return artist_list    
```

0.2. Создаем словарь с мета-инофрмацией о песнях автора

```python
def createsongdict(lyricfile):
    with open(lyricfile, 'r', encoding='utf-8') as f:
        artist_json = json.load(f)
    songs_dict = {}
    for song in artist_json['songs']:
        title = song['title'].replace('\xa0', ' ')
        songs_dict[title] = {'release_date': song['release_date'],
                         'release_date_for_display': song['release_date_for_display'],
                         'lyrics': song['lyrics'].replace('\u2005', ' ').replace('\xa0', ' ')}
    return songs_dict
```

0.3. Функция для того, чтобы найти файл, который скачает api, т.к. genius переименовывает имя исполнителя в свой формат

```python
def find_variations(name, pattern):
    variations = pattern.search(name)
    if variations:
        ru_name = variations[1]
        en_name = variations[2]
        both = ru_name + en_name
        cleaned = re.sub('[\W0-9]', '', both)
        return cleaned
    else:
        cleaned = re.sub('[\W0-9]', '', name)
        return cleaned
```

1. Сама функция, которая создает словарь по списку исполнителей

```python
def scrape(artists):
    api = genius.Genius("5du9ginIKDotA6Jy-86n87gHh_tkGQFtYZpUADSTLZ2SHtxTJ8fEOc5JYphHXZCU")
    pattern = re.compile('(\w+?) \((\w+?)\)')
    artists_dict = {}
    for artist in artists:
        try:
            g_artist = api.search_artist(artist)  # а вот тут интересно! возможно три варианта. 1 - такого исполнителя на сайте нет. 2 - исполнитель на сайте есть, но песен у него нет. 3 - у исполнителя есть песни.
            if g_artist.songs:
                if len(g_artist.songs)>0:
                    genius.remove_section_headers = True
                    genius.skip_non_songs = True
                    genius.excluded_terms = ["(Remix)", "(Live)", "Remix", "mix", "Edition", 'Retouch', '(Retouch)']
                    g_artist.save_lyrics(extension='json', overwrite= True, binary_encoding=True)  #сохранили ВСЮ информацию в json файл
                    name = find_variations(g_artist.name, pattern)  #нашли, как должен называться файл
                    js_file = 'Lyrics_' + name + '.json'  #склеили название файла
                    artist_songs = createsongdict(js_file)
                    artists_dict.update({artist : artist_songs})
                else:
                    with open ('no_data.txt', 'a+', encoding='utf-8') as f:
                        f.write(artist + ' нет песен\n')
            else:
                with open ('no_data.txt', 'a+', encoding='utf-8') as f:
                    f.write(artist + ' нет исполнителя\n')
        except:
            with open ('no_data.txt', 'a+', encoding='utf-8') as f:
                f.write(artist + ' ошибка\n')
    return artists_dict
```

2. Функция, которая сохраняет json файл с данными

```python
def createjson(dictionary, name):
    filename = name + '.json'
    with open(filename, 'w', encoding='utf-8') as f:
        json.dump(dictionary, f, ensure_ascii=False)
```

3. Вся программа

```python
def artist2json(artist_list, filename):
    artists = artistlist(artist_list)
    data = scrape(artists)
    jsonfile = createjson(data, filename)   
```
