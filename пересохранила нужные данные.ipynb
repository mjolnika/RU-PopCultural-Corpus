```python
import json
import os
import re
```

```python
files = os.listdir()
```

```python
json_files = []
for file in files:    
    if file.endswith('.json'):
        json_files.append(file)
```

```python
pattern2 = re.compile('Lyrics_(.+)?\.json')
```

```python
art_dict = {}
for file in json_files:
    artist = pattern2.search(file).group(1)
    with open(file, 'r', encoding='utf-8') as f:
        artist_json = json.load(f)
    songs_dict = {}
    for song in artist_json['songs']:
        try:
            title = song['title'].replace('\xa0', ' ')
            songs_dict[title] = {'release_date': song['release_date'],
                             'release_date_for_display': song['release_date_for_display'],
                             'lyrics': song['lyrics'].replace('\u2005', ' ').replace('\xa0', ' ')}
        except:
            pass
    art_dict.update({artist : songs_dict})
```

```python
def createjson(dictionary, name):
    filename = name + '.json'
    with open(filename, 'w', encoding='utf-8') as f:
        json.dump(dictionary, f, ensure_ascii=False)
```
